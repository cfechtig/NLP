```python
def give_confidence(guess, question_text, file_location):
    count = 0
    repeated = ""
    df = pandas.read_csv(file_location,dtype=str)
    for i in df[guess]:
        if str(i) in question_text and str(i) not in repeated and not str(i) == " ":
            repeated += i
            count += 1
    return count
```
