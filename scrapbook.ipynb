```python
import pandas
import json
```

```python
def make_vocab(qanta_file, out_file):
    qanta = json.load(open(qanta_file))
    vocab_file = open(out_file, "wb")
    vocab = []

    for q in qanta["questions"]:
        x = q["page"].encode('utf8')
        if x not in vocab:
            vocab.append(x)

    for x in vocab:
        vocab_file.write(x + '\n'.encode('utf8'))
    vocab_file.close()

    return vocab
```

```python
train_vocab = make_vocab("qanta.train.2018.04.18.json", "vocab_train.txt")
test_vocab = make_vocab("qanta.test.2018.04.18.json", "vocab_test.txt")
dev_vocab = make_vocab("qanta.dev.2018.04.18.json", "vocab_dev.txt")
```
